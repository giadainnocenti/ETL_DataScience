```python
import pandas as pd
```

```python
# these are the input and output filtered csv files
input_file = "Data/glassdoor_salary_salaries.csv"
output_file = "Data/glassdoor_salary_salaries_filtered.csv"
```

```python
# import the salaries file
salaries_df = pd.read_csv(input_file)
# display the imported file
salaries_df
```

```python
# these are the columns to keep
keep_columns = [
    'id',
    'index',
    #'salary.salaries.val.basePayCount',
    'salary.salaries.val.jobTitle',
    'salary.salaries.val.payPeriod',
    'salary.salaries.val.salaryPercentileMap.payPercentile10',
    'salary.salaries.val.salaryPercentileMap.payPercentile90',
    'salary.salaries.val.salaryPercentileMap.payPercentile50',
    #'salary.salaries.val.salaryType'
]
# the drop columns will be 
drop_columns =  list(salaries_df.columns)
for keep in keep_columns:
    if (keep in drop_columns):
        drop_columns.remove(keep)
# 
print(f"was {len(list(salaries_df.columns))}, keep {len(keep_columns)}, remove {len(drop_columns)}")
```

```python
# drop the unwanted columns
salaries_df = salaries_df.drop(drop_columns, axis=1)
# display the data
salaries_df
```

```python
# drop the rows that are empty at the column
salaries_df = salaries_df[salaries_df["index"].notna()]
# format the saleries from float to ints
salaries_df['index'] = salaries_df['index'].astype(int) 
#
salaries_df
```

```python
# rename the columns to use underscore
columns_rename = list(salaries_df.columns)
for i in range(len(columns_rename)):
    columns_rename[i] = columns_rename[i].replace(".", "_")
    
salaries_df.columns = columns_rename
```

```python
salaries_df
```

```python
# save the csv
salaries_df.to_csv(output_file, index=False)
```
